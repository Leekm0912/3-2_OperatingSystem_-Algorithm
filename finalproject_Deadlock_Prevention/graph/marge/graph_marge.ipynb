```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
circularAtmosphereLog = pd.read_csv("data/"+'CircularAtmosphereLog.txt')
circularAtmosphereLog.columns = ['time', 'running']

holdAndWaitLog = pd.read_csv("data/"+'HoldAndWaitLog.txt')
holdAndWaitLog.columns = ['time', 'running']

nonMutexLog = pd.read_csv("data/"+'NonMutexLog.txt')
nonMutexLog.columns = ['time', 'running']

nonPreemptionLog = pd.read_csv("data/"+'NonPreemptionLog.txt')
nonPreemptionLog.columns = ['time', 'running']
```

```python
plt.figure(figsize=(20,10))
plt.plot(circularAtmosphereLog['time'], circularAtmosphereLog['running'],'r', label = 'circularAtmosphereLog')
plt.plot(holdAndWaitLog['time'], holdAndWaitLog['running'],'g', label='holdAndWaitLog')
plt.plot(nonMutexLog["time"], nonMutexLog["running"], 'b', label='nonMutexLog')
plt.plot(nonPreemptionLog["time"], nonPreemptionLog["running"], 'y', label='nonPreemptionLog')
plt.xlabel('time')
plt.ylabel('running')
plt.yticks(range(0, 160, 10))
plt.legend(fontsize=20)
plt.show()
```
